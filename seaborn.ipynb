# Heatmap: a 2D color-coded matrix that shows relationships or patterns between two variables.
# It's commonly used to:
# - Show correlation between features
# - Visualize confusion matrices -> will see in the later stages.
# - Highlight missing data -> sns.heatmap(df.isnull())

```python
penguins.head()
```

```python
columns=['bill_length_mm', 'bill_depth_mm',	'flipper_length_mm', 'body_mass_g']
```

```python
penguins[columns].corr()
```

# postiove correaltion means if one increase other also increases

```python
sns.heatmap(data=penguins[columns].corr(), annot =True,cmap='coolwarm', vmin=-0.2,linewidth="2", linecolor="black")#annot gives the numeric values,vmin means from where staring the values
```

# pair grid -> subplot grid for plotting pairwise relationships in a dataset.

```python

graph = sns.PairGrid(data=penguins, hue="sex", palette="Set2")

graph.map_upper(sns.scatterplot) #upper triangel
graph.map_lower(sns.kdeplot) #lower triangle
graph.map_diag(sns.histplot) #diagonal
graph.add_legend()
```
