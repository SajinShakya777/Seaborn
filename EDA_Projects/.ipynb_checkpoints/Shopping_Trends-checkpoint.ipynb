#### Basic Syntax of pivot table
The primary function is pd.pivot_table(), which requires four main components:

1.data: The input DataFrame.index: The column(s) to group by on the rows.

2.columns: The column(s) to group by on the columns.

3.values: The specific column(s) you want to aggregate.

4.aggfunc: The mathematical operation to perform (e.g., 'mean', 'sum', 'count'). The default is 'mean'

```python
pd.pivot_table(data, values = "Purchase Amount (USD)", index ="Location", columns= "Item Purchased", aggfunc="mean")
```

```python
pd.pivot_table(data, values = "Purchase Amount (USD)", index ="Location", columns= "Item Purchased", aggfunc="mean").head()
```
