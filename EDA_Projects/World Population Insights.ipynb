```python
sns.histplot(df['2023 population'], bins = 10, kde= True)
plt.title('distribution of 2023 population')
plt.xlabel('population')
plt.ylabel('frequency')
plt.show()
    
```

```python
df.groupby('continent')['2023 population'].mean()
```

```python
#return to panel dataset
df.groupby('continent')['2023 population'].mean().reset_index()
```

```python
df.groupby('continent')['2023 population'].mean().sort_values(ascending = False)
```

```python
df.groupby('continent')['2023 population'].mean().sort_values(ascending = False).head(2)
```

```python
top_10_countries = df.sort_values(by ='2023 population', ascending = False).head(10)
```

```python
top_10_countries
```

```python
sns.barplot(x='2023 population',y ='country', data = top_10_countries, hue='2023 population', palette='viridis')
plt.title('top_10_countries by population in 2023')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
df.columns
```

```python
 df.rename(columns={'density (km²)': 'density(km2)'}, inplace = True)
```

```python
df.columns
```

```python
sns.scatterplot(x='2023 population', y = 'density(km2)', data = df)
plt.title("Population and density scatter plot")
plt.ylabel("density ")
plt.xlabel("2023 population")
plt.show()
```

```python
df.groupby('continent')['area (km²)'].sum()
```

```python
df.groupby('continent')['area (km²)'].sum().sort_values(ascending= False)
```

```python
sns.violinplot(x='continent',y='density(km2)',data=df,palette='viridis',hue='continent',legend=False)

plt.title('Population Density Distribution Across Continents')
plt.xlabel('Continents')
plt.ylabel('Population Density')
plt.show()
```

```python
sns.boxplot(x = "continent", y = "density(km2)", data = df, palette = "muted")

plt.title("Population Density Distribution Across Continents (Without Outliers)")
plt.xlabel("Continent")
plt.ylabel("Population Density")
plt.show()
```

```python
sns.boxplot(
    x="continent",
    y="density(km2)",
    data=df,
    hue="continent",
    palette="muted",
    legend=False,
    showfliers=False
)

plt.title("Population Density Distribution Across Continents (Without Outliers)")
plt.xlabel("Continent")
plt.ylabel("Population Density")
plt.show()
```
